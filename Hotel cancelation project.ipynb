# Data analysis and visualization

```python
# now we check the total no of cancelation and not cancelation with respect to total no of cancelation
canceled_per = df['is_canceled'].value_counts(normalize=True)
canceled_per
```

```python
# now we make the bar chart to showcase the number of canclation with respect to not canceled booking
plt.figure(figsize=(5,4))
plt.title("Reservation Status Count")
plt.bar(["Not Canceled","canceled"],df['is_canceled'].value_counts(),edgecolor="k",width=0.7)
plt.grid(False)
plt.show()
```

# so her we can see that canceled reservation is 37% which is preety much higher so we cant neglect it

```python
# now will make a bar chart for different hotel type and check the number of booking being canceled with respect to 
#booking not canceled

plt.figure(figsize=(8,4))
sns.countplot(data=df,x="hotel",hue="is_canceled",palette="Set1")
plt.title("Reservation Status in different hotel",size=20)
plt.xlabel("hotel")
plt.ylabel("count")
plt.legend("booking Staus",labels=["not canceled","canceled"])
plt.show()

```

## so here we can see that city hotel have more booking as well as more cancelation than the resort hotel

```python
# first we will check the number of cancelation for the resort hotel
resort_hotel= df[df['hotel']=="Resort Hotel"]
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
# first we will check the number of cancelation for the city hotel
city_hotel= df[df['hotel']=="City Hotel"]
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
# now we check the adr for every reservation date for different type of hotel
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# plot the line chart for both the resort and city hotel to check the adr of of every reservation date
plt.figure(figsize=(20, 8))

plt.title("Average daily rate in city and resort hotel", fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.grid(False)
plt.show()
```

# In order to get less booking for resort hotel as compared to city hotel we check the adr for city hotel and resort hotel 

```python
df['month'] = df['reservation_status_date'].dt.month

# Create the bar plot.
plt.figure(figsize=(16, 8))
sns.set(style="whitegrid")
sns.countplot(data=df, x='month', hue='is_canceled', palette="bright")

plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Canceled vs. Not Canceled Bookings by Month',size=20)
plt.legend(title='Booking Status', labels=['Canceled', 'Not Canceled'])
plt.grid(False)
plt.show()
```

# Here we can see that august is month of larger reserevation whereas january is most canceled reservation

```python
# now we check the if adr making t an effect on cancelation  in every month for the hotel
```

```python
plt.figure(figsize=(15,6))
plt.title("Adr per month",fontsize=20)
sns.barplot(x='month',y='adr', data=df[df['is_canceled']==1].groupby('month')['adr'].sum().reset_index())
plt.show()
```

# So it is clear that adr making the effect on cancelation january has the  highest canclation and also the adr of the january is also the highest
